```python
def gap(df):
    """Returns a dataframe with a column of business days and the difference between 
    the closing of and opening stock prices of consecutive days
    
    Example output:
           Date_Time   Gap
        0 2016-10-10   NaN
        1 2016-10-11 -0.38
        2 2016-10-12 -0.09
        3 2016-10-13 -0.71
        4 2016-10-14  0.41
    """
    PERIODS_PER_DAY = 130
    grouped = (df.shift(-PERIODS_PER_DAY).resample('B')['Open'].first() - # The opening price of the day
               df.resample('B')['Close'].last()).shift()                  # The closing price of the next day
    grouped = grouped.reset_index().rename(columns={0: 'Gap'})
    return grouped
```

```python
def highest_highs(df, periods=30):
    """Returns the highest value from the 'High' column over the next consecutive *periods*"""
    return df.shift(-periods).rolling(periods)['High'].max().rename("Highest High")
```

```python
def lowest_lows(df, periods=30):
    """Returns the lowest value from the 'Low' column over the next consecutive *periods*"""
    return df.shift(-periods).rolling(periods)['Low'].min().rename("Lowest Low")
```

```python
def trend(df):
    """Returns dataframe after added columns for highest highs, lowest lows, and the differences
    between these values and the closing price for each row."""
    high = highest_highs(df)
    low = lowest_lows(df)
    df = df.join([high, low])
    df['High Difference'] = np.abs(df['Close'] - df['Highest High'])
    df['Low Difference'] = np.abs(df['Close'] - df['Lowest Low'])
    return df
```

```python
# Add four new columns (as described in the trend() function)
df = trend(df)
```

```python
def compute_target(diff, delta_price=0.10):
    """Use with pd.apply to compute target"""
    if abs(diff) < delta_price:
        return 'FLAT'
    else:
        if diff < 0:
            return 'DOWN'
        elif diff > 0:
            return 'UP'
        else:
            return 'NaN'
```

```python
# Add a target column
df['Target'] = df['High Difference'].sub(df['Low Difference']).apply(compute_target)
```

```python
def max_high_first_30_minutes(df):
    """
    Args:
        df: A pandas dataframe with a datetime multi-index
            The first level is a business day, and the
            second level is the first 30 minutes of the day
            
    Returns:
        A series with business day index and maximum high
        from the first 30 minutes of the day.
    """
    return df.groupby(pd.Grouper(freq='B', level=0)).High.max()
```

```python
def min_low_first_30_minutes(df):
    """
    Args:
        df: A pandas dataframe with a datetime multi-index
            The first level is a business day, and the
            second level is the first 30 minutes of the day
            
    Returns:
        A series with business day index and minimum low
        from the first 30 minutes of the day.
    """
    return df.groupby(pd.Grouper(freq='B', level=0)).Low.min()
```

```python
# Is the gap negative? If so, did the price trend down in the first 30 minutes of the day?
def downward_continuation(first30_with_gap):
    negative_gaps = np.sign(first30_with_gap.Gap) == -1
    downward_trends = np.sign(first30_with_gap.First30_Open - first30_with_gap.First30_Close) == -1
    return negative_gaps & downward_trends
```

```python
# The price went down in the first 30 minutes (ten periods) after a negative gap about 19% of the time
down_cont = downward_continuation(first30_with_gap)
down_cont.mean()
```

```python
# Given there was a negative gap, the first bar was DOWN with probability:
np.sum(down_cont) / np.sum(direction_of_gaps == -1)
```

```python
# Is the gap positive? If so, did the price trend up in the first 30 minutes of the day?
def upward_continuation(first30_with_gap):
    positive_gaps = np.sign(first30_with_gap.Gap) == 1
    upward_trends = np.sign(first30_with_gap.First30_Open - first30_with_gap.First30_Close) == 1
    return positive_gaps & upward_trends
```

```python
# The price went up in the first 30 minutes (ten periods) after a positive gap about 27% of the time
up_cont = upward_continuation(first30_with_gap)
up_cont.mean()
```

```python
# Given there was a positive gap, the first bar was UP with probability:
np.sum(up_cont) / np.sum(direction_of_gaps == 1)
```

```python
def count_first_consecutive_after_negative_gap(x):
    gap_sign = np.sign(np.mean(x['Gap']))
    if gap_sign == -1:
        a = (x['Target'] == 'DOWN').values
    else:
        return np.nan
    return np.argmin(np.append(a[~np.logical_and.accumulate(~a)], False))
```

```python
def count_first_consecutive_after_positive_gap(x):
    gap_sign = np.sign(np.mean(x['Gap']))
    if gap_sign == 1:
        a = (x['Target'] == 'UP').values
    else:
        return np.nan
    return np.argmin(np.append(a[~np.logical_and.accumulate(~a)], False))
```
